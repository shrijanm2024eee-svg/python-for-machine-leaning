```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([5, 7, 2, 15])

plt.bar(x,y)
plt.show()
```

```python
#Horizontal bar
plt.barh(x,y)
plt.show()
```

```python
#Color option
plt.bar(x,y, color='red')
plt.show()
```

```python
#width option
plt.bar(x,y, color='g', width=0.1)
plt.show()
```

```python
#Histogram
x = np.random.normal(300, 20, 250)
plt.hist(x)
plt.show()
```

```python
#Pie chart
y = np.array([40, 30, 20, 10])
plt.pie(y)
plt.show()
```

```python
#With Labels
y = np.array([35, 25, 25, 15])
mylabels = ["MPMC", "PPML", "ContolSystem", "PSA"]
plt.pie(y, labels = mylabels)
plt.show()
```

```python
#Explode
y = np.array([35, 25, 25, 15])
mylabels = ["MPMC", "PPML", "ContolSystem", "PSA"]

myexplode = [0, 0, 0.3, 0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```
